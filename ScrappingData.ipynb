## Scrapping Dataset Review Aplikasi TheoTown di Google Play Store

**Import Library**

```python
!pip install google-play-scraper
```

```python
# import google_play_scrapper
from google_play_scraper import app, reviews, Sort, reviews_all
```

```python
# Mengambil semua review dari aplikasi TheoTown di Google Play Store.
ulasan = reviews_all(
    'info.flowersoft.theotown.theotown',
    lang='id',
    country='id',
    sort=Sort.MOST_RELEVANT,
    count=15000          #15rb data
)
```

**Simpan ke csv**

```python
import pandas as pd
```

```python
# Simpan review ke csv
df = pd.DataFrame(ulasan)
df.to_csv('theotown_reviews.csv', index=False)
```
